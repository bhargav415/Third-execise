```python
''' 1. Write a Python function to count the no. of occurances of 4 from the data given below.
    Data:
        [1, 4, 6, 7, 4]
        [1, 4, 6, 4, 7, 4]
    Expected Output:
        2
        3 '''

```

```python
def count_occurrences(data, target):
    count = 0
    for num in data:
        if num == target:
            count += 1
    return count

data1 = [1, 4, 6, 7, 4]
data2 = [1, 4, 6, 4, 7, 4]

count1 = count_occurrences(data1, 4)
count2 = count_occurrences(data2, 4)

print(count1)
print(count2)
```

```python
''' 6. Write a python function to print the sum of digits for the given data and print the execution time (in seconds).
        Data:
            n = 5
        Expected Output:
            Time to sum of 1 to 5 and required time to calculate is : (15, 0.0) '''
            
```

```python
import time
 
def sum_1_to_n_with_time(n):
    start = time.time()
    total = sum(range(1, n+1))
    end = time.time()
    elapsed = end - start
    print(f"Time to sum of 1 to {n} and required time to calculate is : ({total}, {elapsed})")
 
# Example usage
n = 5
sum_1_to_n_with_time(n)
```

```python
''' 7. Write a Python program to hash a word.
        Data:
            x = [0,1,2,3,0,1,2,0,0,2,2,4,5,5,0,1,2,6,2,3,0,1,0,2,0,2]
            word = Input the word be hashed: rfdf
        Expected Output:
            Enter the word to be hashed: culture
            The coded word is: C210220 '''

```

```python
def hash_word(word, data):
    result = ""
    for char in word.upper():
        if 'A' <= char <= 'Z':
            index = ord(char) - ord('A')
            result += str(data[index % len(data)])
        else:
            result += char  # Keep non-alphabetic characters as they are
    return result
data = [0, 1, 2, 3, 0, 1, 2, 0, 0, 2, 2, 4, 5, 5, 0, 1, 2, 6, 2, 3, 0, 1, 0, 2, 0, 2]
word = input("Enter the word to be hashed: ")
hashed_word = hash_word(word, data)
print(f"The coded word is: {hashed_word}")
```

```python
''' 9. Write a Python program to calculate the time runs (difference between start and current time) of a program.
        Data:
            times(5)
            times(4)
        Expected Output:
            0
            1
            2
            3
            4
            0.003652000000101907
            0
            1
            2
            3
            0.0070380999998178595 '''
```

```python
import time
 
def times(n):
    start = time.perf_counter()
    for i in range(n):
        print(i)
    elapsed = time.perf_counter() - start
    print(elapsed)
times(5)
times(4)
```

```python
''' 10. Write a Python function to find the maximum and minimum numbers from the given list.
        Data:
            seq = [0, 10, 15, 40, -5, 42, 17, 28, 75]
        Expected Output:
            (75, -5) '''
```

```python
def find_max_min(data):
    if not data:
        return (None, None)  
    max_num = data[0]
    min_num = data[0]

    for num in data:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num

    return (max_num, min_num)
seq = [0, 10, 15, 40, -5, 42, 17, 28, 75]
max_val, min_val = find_max_min(seq)
print(f"({max_val}, {min_val})")
```

```python
''' 11. Write a Python function to find a distinct pair of numbers whose product is odd from the given data. Print True if pair found, none if not.
        Data:
            x = [2, 4, 6, 8]
            y = [1, 6, 4, 7, 8]
        Expected Output:
            None
            True '''
```

```python
def odd_product(nums):
    for i in range(len(nums)):
        for j in range(len(nums)):
            if i != j:
                product = nums[i] * nums[j]
                if product & 1:
                    return True
    return False
dt1 = [2, 4, 6, 8]
dt2 = [1, 6, 4, 7, 8]
dt3 = [1, 3, 5, 7, 9]

print(dt1, odd_product(dt1))
print(dt2, odd_product(dt2))
print(dt3, odd_product(dt3))
```

```python
import itertools

def generate_permutations(numbers):
    return list(itertools.permutations(numbers))
numbers = [1, 2, 3]
permutations = generate_permutations(numbers)
for perm in permutations:
    print(list(perm))
```

```python
''' 14. Write a Python program to find the number of notes (Sample of notes: 10, 20, 50, 100, 200 and 500).
        Data:
            880
            1000
        Expected Output:
            6
            2 '''
```

```python
def count_notes(amount, notes):
    note_counts = {}
    for note in notes:
        if amount >= note:
            num_notes = amount // note
            note_counts[note] = num_notes
            amount %= note
    if amount == 0:
        return note_counts
    else:
        return None  
available_notes = [500, 200, 100, 50, 20, 10]
amounts_to_process = [880, 1000]

for amount in amounts_to_process:
    result = count_notes(amount, available_notes)
    if result is not None:
        total_notes = sum(result.values())
        print(f"For amount {amount}: Total notes = {total_notes}, Breakdown: {result}")
    else:
        print(f"For amount {amount}: Cannot fully dispense with available notes.")
```

```python
''' 15. Write a Python function to find the number of divisors of a given integer is even or odd.
        Data:
            x = 15, x = 12, x = 9, x = 6, x = 3
        Expected Output:
            4
            6
            3
            4
            2 '''
```

```python
def count_divisors(x):
    count = 0
    for i in range(1, x + 1):
        if x % i == 0:
            count += 1
    return count
 
# Test data
for x in [15, 12, 9, 6, 3]:
    print(count_divisors(x))
```

```python
''' 18. Write a Python function to multiply all the items in a list.
        Data:
            list = [1, 2, -8]
        Expected Output:
            -16 '''
```

```python
def multiply_list(data):
    product = 1
    for item in data:
        if isinstance(item, str):
            try:
                item = int(item)
            except ValueError:
                return "Error: Invalid input. List contains non-numeric values."
        product *= item
    return str(product)

my_list = [1, 2, "-8"]
result = multiply_list(my_list)
print(result)
```

```python
''' 20. Write a Python program to remove duplicates from the data given below.
        Data:
            a = [10,20,30,20,10,50,60,40,80,50,40]
        Expected Output:
            [10, 20, 30, 50, 60, 40, 80] '''
```

```python
a = [10, 20, 30, 20, 10, 50, 60, 40, 80, 50, 40]

dup_items = set()
uniq_items = []

for x in a:
    if x not in dup_items:
        uniq_items.append(x)
        dup_items.add(x)
print(dup_items) 
```
